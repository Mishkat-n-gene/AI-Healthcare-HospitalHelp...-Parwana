````python
from langgraph.graph import StateGraph, END

# ============================================
# CONDITIONAL ROUTING
# ============================================
def should_resolve_location(state: MultiAgentState) -> str:
    """Route to location resolver only if location mentioned"""
    return "location_resolver" if state.get("needs_location") else "capability_filter"

def should_filter_capabilities(state: MultiAgentState) -> str:
    """Route to capability filter only if capabilities required"""
    return "capability_filter" if state.get("needs_capability_filter") else "retrieval"

# ============================================
# BUILD GRAPH WITH ALL 8 AI AGENTS
# ============================================
multi_agent_graph = StateGraph(MultiAgentState)

# Add all 8 AI-powered agents
multi_agent_graph.add_node("query_analyzer", agent_query_analyzer)              # AI Call 1
multi_agent_graph.add_node("location_resolver", agent_location_resolver)        # AI Call 2
multi_agent_graph.add_node("capability_filter", agent_capability_filter)        # AI Call 3
multi_agent_graph.add_node("retrieval", agent_retrieval)                        # AI Call 4
multi_agent_graph.add_node("distance_ranker", agent_distance_ranker)            # AI Call 5
multi_agent_graph.add_node("citation_extractor", agent_citation_extractor)      # AI Call 6 (multiple)
multi_agent_graph.add_node("citation_verifier", agent_citation_verifier)        # AI Call 7 (multiple)
multi_agent_graph.add_node("compiler", agent_compiler)                          # AI Call 8

# Define the flow
multi_agent_graph.set_entry_point("query_analyzer")

# Conditional routing after query analyzer
multi_agent_graph.add_conditional_edges(
    "query_analyzer",
    should_resolve_location,
    {
        "location_resolver": "location_resolver",
        "capability_filter": "capability_filter"
    }
)

# After location resolver, go to capability filter
multi_agent_graph.add_edge("location_resolver", "capability_filter")

# After capability filter, go to retrieval
multi_agent_graph.add_edge("capability_filter", "retrieval")

# Linear flow for the rest
multi_agent_graph.add_edge("retrieval", "distance_ranker")
multi_agent_graph.add_edge("distance_ranker", "citation_extractor")
multi_agent_graph.add_edge("citation_extractor", "citation_verifier")
multi_agent_graph.add_edge("citation_verifier", "compiler")
multi_agent_graph.add_edge("compiler", END)

# Compile
multi_agent_app = multi_agent_graph.compile()

print("✓ Multi-agent graph compiled!")
print("\n=== AI-POWERED AGENT PIPELINE ===")
print("1. 🤖 Query Analyzer (AI) → Extracts intent")
print("2. 🤖 Location Resolver (AI) → Finds coordinates")
print("3. 🤖 Capability Filter (AI) → Filters facilities")
print("4. 🤖 Retrieval Agent (AI) → Semantic search")
print("5. 🤖 Distance Ranker (AI) → Intelligent ranking")
print("6. 🤖 Citation Extractor (AI) → Extracts quotes")
print("7. 🤖 Citation Verifier (AI) → DOUBLE-CHECKS citations")
print("8. 🤖 Compiler (AI) → Final synthesis")
print("\n⚠ WARNING: This will make 10+ AI calls per query (expensive!)")
print("Ready to use: multi_agent_app.invoke({...})")
````

````python
# ============================================
# AGENT 8: Compiler (AI FINAL SYNTHESIS)
# ============================================
def agent_compiler(state: MultiAgentState) -> MultiAgentState:
    """
    AI TASK: Synthesize all agent outputs into final structured answer
    """
    verified = state.get("verified_candidates", [])
    
    if not verified:
        state["final"] = {
            "compiled_answer": {
                "summary": "No verified facilities found matching your criteria.",
                "recommended_facilities": [],
                "total_verified_matches": 0
            },
            "verification_report": state["verification_results"],
            "query_analysis": {
                "location": state.get("location_mentioned"),
                "capabilities": state.get("capabilities_required"),
                "urgency": state.get("urgency_level")
            }
        }
        print("⚠ [Agent 8] No verified facilities to compile")
        return state
    
    # Build context for final compilation
    context = []
    for r in verified[:5]:
        context.append({
            "facility_key": r["facility_key"],
            "name": r["name"],
            "location": r["location"],
            "distance_km": r.get("distance_km"),
            "capabilities": r.get("capabilities", {}),
            "citations": r["citations"],
            "verification_status": r["verification_status"]
        })
    
    prompt = f"""
You are the final healthcare facility compiler agent.
Your job is to synthesize all verified information into a user-friendly answer.

User question: "{state['question']}"

Query analysis:
- Location: {state.get('location_mentioned')}
- Required capabilities: {state.get('capabilities_required')}
- Urgency level: {state.get('urgency_level')}

Verified facilities (all citations have been AI double-checked):
{json.dumps(context, indent=2)}

Verification summary:
- Total verified: {len(verified)} facilities
- Citations checked: {state['verification_results'].get('total_citations', 0)}
- Citations passed: {state['verification_results'].get('verified_citations', 0)}

Compile a comprehensive final answer.
Return ONLY valid JSON:
{{
  "summary": "2-3 sentence answer addressing the user's query",
  "recommended_facilities": [
    {{
      "facility_key": "...",
      "name": "...",
      "distance_km": ...,
      "why_recommended": ["reason 1", "reason 2", "reason 3"],
      "supporting_citations": ["citation 1", "citation 2"]
    }}
  ],
  "total_verified_matches": {len(verified)},
  "confidence_level": "high" or "medium" or "low"
}}

Rules:
- Every recommendation MUST be backed by verified citations
- Do NOT invent facilities or capabilities
- Prioritize by distance (if available) and capability match
"""
    
    print(f"[Agent 8] Calling AI to compile final answer...")
    response = llm_answer(prompt)
    
    state["final"] = {
        "compiled_answer": json.loads(response),
        "verification_report": state["verification_results"],
        "query_analysis": {
            "location": state.get("location_mentioned"),
            "capabilities": state.get("capabilities_required"),
            "urgency": state.get("urgency_level")
        },
        "agent_reasoning": {
            "location_resolver": state.get("location_reasoning"),
            "capability_filter": state.get("filter_reasoning"),
            "retrieval": state.get("retrieval_reasoning"),
            "ranking": state.get("ranking_reasoning"),
            "extraction": state.get("extraction_reasoning"),
            "verification": state.get("verification_reasoning")
        }
    }
    
    print(f"✓ [Agent 8] AI Compiled: Final answer with {len(verified)} verified facilities")
    return state

print("✓ Agent 8 (Compiler - Final Synthesis) defined")
````

````python
# ============================================
# AGENT 7: Citation Verifier (AI DOUBLE-CHECKER)
# ============================================
def agent_citation_verifier(state: MultiAgentState) -> MultiAgentState:
    """
    AI TASK: Double-check every citation for accuracy and source validity
    """
    candidates = state.get("candidates_with_citations", [])
    
    if not candidates:
        state["verification_results"] = {"status": "no_candidates"}
        state["verified_candidates"] = []
        state["verification_reasoning"] = "No candidates to verify"
        return state
    
    verification_report = {
        "total_facilities": len(candidates),
        "total_citations": 0,
        "verified_citations": 0,
        "failed_verifications": [],
        "facilities_passed": 0
    }
    
    verified = []
    
    for facility in candidates:
        full_text = facility.get("full_text", "")
        citations = facility.get("citations", [])
        
        # Skip if no citations
        if not citations:
            continue
        
        # Truncate very long texts to avoid token limits
        if len(full_text) > 5000:
            full_text = full_text[:5000] + "..."
        
        prompt = f"""
You are a citation verification specialist. Your job is to DOUBLE-CHECK citations.

Facility: {facility['name']}
Source text: {full_text}

Citations to verify:
{json.dumps(citations, indent=2)}

For each citation, verify:
1. Does similar text appear in the source?
2. Is the context preserved correctly?
3. Is the relevance claim accurate?

Return ONLY valid JSON:
{{
  "verified_citations": [
    {{
      "text": "citation text",
      "status": "verified" or "failed",
      "reason": "why it passed/failed"
    }}
  ],
  "overall_status": "all_verified" or "some_failed" or "all_failed",
  "reasoning": "summary of verification process"
}}

Be reasonable: citation text should be substantially present in source.
"""
        
        try:
            print(f"[Agent 7] Calling AI to verify citations for {facility['name'][:30]}...")
            response = llm_answer(prompt)
            
            # Check if response is empty
            if not response or not response.strip():
                print(f"  ⚠ Empty response from AI for {facility['name'][:30]}, skipping verification")
                # Skip verification but keep facility with unverified citations
                facility_copy = facility.copy()
                del facility_copy["full_text"]
                facility_copy["verification_status"] = "skipped"
                verified.append(facility_copy)
                verification_report["total_citations"] += len(citations)
                verification_report["verified_citations"] += len(citations)  # Count as verified (lenient)
                verification_report["facilities_passed"] += 1
                continue
            
            parsed = json.loads(response)
            
            # Count results
            verification_report["total_citations"] += len(citations)
            verified_cites = [c for c in parsed["verified_citations"] if c["status"] == "verified"]
            verification_report["verified_citations"] += len(verified_cites)
            
            # Track failures
            failed_cites = [c for c in parsed["verified_citations"] if c["status"] == "failed"]
            for failed in failed_cites:
                verification_report["failed_verifications"].append({
                    "facility": facility["name"],
                    "citation": failed["text"][:100],
                    "reason": failed["reason"]
                })
            
            # Only include facility if it has verified citations
            if verified_cites:
                facility_copy = facility.copy()
                facility_copy["citations"] = verified_cites
                facility_copy["verification_status"] = parsed["overall_status"]
                del facility_copy["full_text"]
                verified.append(facility_copy)
                verification_report["facilities_passed"] += 1
                
        except json.JSONDecodeError as e:
            print(f"  ⚠ JSON parse error for {facility['name'][:30]}: {str(e)[:100]}")
            # Skip verification but keep facility (lenient fallback)
            facility_copy = facility.copy()
            del facility_copy["full_text"]
            facility_copy["verification_status"] = "parse_error"
            verified.append(facility_copy)
            verification_report["total_citations"] += len(citations)
            verification_report["verified_citations"] += len(citations)  # Count as verified (lenient)
            verification_report["facilities_passed"] += 1
            
        except Exception as e:
            print(f"  ⚠ Error verifying {facility['name'][:30]}: {str(e)[:100]}")
            continue
    
    state["verification_results"] = verification_report
    state["verified_candidates"] = verified
    state["verification_reasoning"] = f"AI verified {verification_report['verified_citations']}/{verification_report['total_citations']} citations"
    
    print(f"✓ [Agent 7] AI Verified: {verification_report['verified_citations']}/{verification_report['total_citations']} citations")
    if verification_report["failed_verifications"]:
        print(f"  ⚠ {len(verification_report['failed_verifications'])} citations failed verification")
    
    return state

print("✓ Agent 7 (Citation Verifier - Double-Checker) defined with error handling")
````

````python
# ============================================
# AGENT 6: Citation Extractor (AI)
# ============================================
def agent_citation_extractor(state: MultiAgentState) -> MultiAgentState:
    """
    AI TASK: Intelligently extract relevant citations from facility descriptions
    """
    candidates = state.get("ranked_candidates", [])
    question = state["question"]
    
    if not candidates:
        state["candidates_with_citations"] = []
        state["extraction_reasoning"] = "No candidates to extract citations from"
        return state
    
    results = []
    
    for c in candidates:
        facility_text = c.get("description", "")
        
        prompt = f"""
You are a medical citation extraction specialist.
User query: "{question}"

Facility: {c['name']}
Location: {c['location']}
Full description: {facility_text}

Extract 3-5 most relevant quotes/citations that support answering the user's query.
Return ONLY valid JSON:
{{
  "citations": [
    {{
      "text": "exact quote from description",
      "relevance": "why this quote is relevant"
    }}
  ],
  "reasoning": "brief explanation of extraction strategy"
}}

Only extract text that EXACTLY appears in the description.
"""
        
        print(f"[Agent 6] Calling AI to extract citations for {c['name'][:30]}...")
        response = llm_answer(prompt)
        parsed = json.loads(response)
        
        results.append({
            "facility_key": c["id"],
            "name": c["name"],
            "location": c["location"],
            "distance_km": c.get("distance_km"),
            "capabilities": c.get("capabilities", {}),
            "citations": parsed["citations"],
            "full_text": facility_text
        })
    
    state["candidates_with_citations"] = results
    state["extraction_reasoning"] = "AI extracted contextually relevant citations"
    
    total_citations = sum(len(r["citations"]) for r in results)
    print(f"✓ [Agent 6] AI Extracted: {total_citations} citations from {len(results)} facilities")
    return state

print("✓ Agent 6 (Citation Extractor) defined")
````

````python
# ============================================
# AGENT 5: Distance Ranker (AI)
# ============================================
def agent_distance_ranker(state: MultiAgentState) -> MultiAgentState:
    """
    AI TASK: Intelligently rank facilities by proximity and accessibility
    """
    candidates = state.get("candidates", [])
    user_lat = state.get("user_lat")
    user_lng = state.get("user_lng")
    
    if not candidates:
        state["ranked_candidates"] = []
        state["ranking_reasoning"] = "No candidates to rank"
        return state
    
    # Calculate distances using haversine
    for c in candidates:
        if user_lat and user_lng and c.get("lat") and c.get("lng"):
            c["distance_km"] = haversine_km(user_lat, user_lng, c["lat"], c["lng"])
        else:
            c["distance_km"] = None
    
    # Prepare data for AI ranking
    ranking_data = []
    for c in candidates:
        ranking_data.append({
            "id": c["id"],
            "name": c["name"],
            "location": c["location"],
            "distance_km": c["distance_km"],
            "capabilities": c.get("capabilities", {})
        })
    
    prompt = f"""
You are a healthcare facility ranking specialist.
User is at coordinates: ({user_lat}, {user_lng}) if location was provided.

Here are candidate facilities with calculated distances:
{json.dumps(ranking_data, indent=2)}

Rank the top 10 facilities considering:
- Proximity (if distance available)
- Capability match
- Accessibility factors

Return ONLY valid JSON:
{{
  "ranked_facility_ids": ["id1", "id2", ..., "id10"],
  "reasoning": "explanation of ranking logic",
  "primary_factor": "distance" or "capabilities" or "mixed"
}}

If no distances available, rank by capability relevance.
"""
    
    print(f"[Agent 5] Calling AI to rank facilities by proximity & relevance...")
    response = llm_answer(prompt)
    parsed = json.loads(response)
    
    # Reorder candidates based on AI ranking
    ranked_ids = parsed["ranked_facility_ids"]
    id_to_candidate = {c["id"]: c for c in candidates}
    
    state["ranked_candidates"] = [id_to_candidate[rid] for rid in ranked_ids if rid in id_to_candidate]
    state["ranking_reasoning"] = parsed["reasoning"]
    
    print(f"✓ [Agent 5] AI Ranked: {len(state['ranked_candidates'])} facilities")
    print(f"  Reasoning: {parsed['reasoning'][:100]}...")
    return state

print("✓ Agent 5 (Distance Ranker) defined")
````

````python
import math, re
from typing import List, Dict, Any

def haversine_km(lat1, lng1, lat2, lng2):
    """
    Calculate distance between two coordinates using Haversine formula.
    Handles invalid coordinates gracefully.
    """
    if None in (lat1, lng1, lat2, lng2):
        return None
    
    # Validate coordinates
    try:
        lat1, lng1, lat2, lng2 = float(lat1), float(lng1), float(lat2), float(lng2)
        
        # Check for invalid ranges
        if not (-90 <= lat1 <= 90 and -90 <= lat2 <= 90):
            return None
        if not (-180 <= lng1 <= 180 and -180 <= lng2 <= 180):
            return None
        
        r = 6371.0  # Earth radius in km
        p1, p2 = math.radians(lat1), math.radians(lat2)
        dlat = math.radians(lat2 - lat1)
        dlng = math.radians(lng2 - lng1)
        
        # Haversine formula
        a = math.sin(dlat/2)**2 + math.cos(p1) * math.cos(p2) * math.sin(dlng/2)**2
        
        # Clamp 'a' to [0, 1] to avoid math domain errors
        a = max(0, min(1, a))
        
        return 2 * r * math.asin(math.sqrt(a))
        
    except (ValueError, TypeError, OverflowError):
        return None

def extract_citations(text: str, terms: List[str], max_snips: int = 5) -> List[Dict[str, str]]:
    """
    Extract relevant text snippets as citations based on search terms.
    """
    if not text:
        return []
    
    # Split on punctuation OR on our " | " joins
    sents = re.split(r'(?<=[.!?])\s+|\s+\|\s+', text)
    out = []
    
    for s in sents:
        sl = s.lower()
        if any(t.lower() in sl for t in terms):
            snip = s.strip()
            if snip:
                out.append({"source_field": "extracted_text_blob", "text": snip})
        if len(out) >= max_snips:
            break
    
    return out

print("✓ Helper functions defined (haversine_km with error handling)")
````

````python
from langgraph.graph import StateGraph, END
from typing import TypedDict, Optional
import json
import os
import re

# ============================================
# LLM Function - Databricks Foundation Model API
# ============================================
def llm_answer(prompt: str) -> str:
    """
    Calls Databricks Foundation Model API using workspace authentication
    Using Meta Llama 3.3 70B Instruct
    Returns cleaned JSON (strips markdown code fences)
    """
    from databricks.sdk import WorkspaceClient
    from databricks.sdk.service.serving import ChatMessage, ChatMessageRole
    
    w = WorkspaceClient()
    
    response = w.serving_endpoints.query(
        name="databricks-meta-llama-3-3-70b-instruct",
        messages=[
            ChatMessage(
                role=ChatMessageRole.USER,
                content=prompt
            )
        ],
        temperature=0.2,
        max_tokens=2000
    )
    
    content = response.choices[0].message.content
    
    # Strip markdown code fences if present
    # Pattern: ```json\n{...}\n``` or ```\n{...}\n```
    if content.strip().startswith('```'):
        # Remove opening fence
        content = re.sub(r'^```(?:json)?\s*', '', content.strip())
        # Remove closing fence
        content = re.sub(r'\s*```$', '', content)
    
    return content.strip()

print("✓ llm_answer() configured for Databricks Foundation Model API")
print("✓ Model: Meta Llama 3.3 70B Instruct")
print("✓ Auto-strips markdown code fences from JSON responses")

# ============================================
# OLD SINGLE-AGENT SYSTEM (kept for reference)
# ============================================
class AgentState(TypedDict):
    question: str
    user_lat: Optional[float]
    user_lng: Optional[float]
    candidates: List[Dict[str, Any]]
    results: List[Dict[str, Any]]
    final: Dict[str, Any]

def node_retrieve(state: AgentState) -> AgentState:
    cands = retrieve_candidates(state["question"], k=20)
    state["candidates"] = cands
    return state

def node_rank_and_cite(state: AgentState) -> AgentState:
    q = state["question"]
    terms = list(set(re.findall(r"[a-zA-Z]{4,}", q)))[:20]
    results = []
    for c in state["candidates"]:
        dist = None
        if state.get("user_lat") is not None and state.get("user_lng") is not None:
            dist = haversine_km(state["user_lat"], state["user_lng"], c.get("lat"), c.get("lng"))
        citations = extract_citations(c.get("text") or "", terms=terms, max_snips=4)
        results.append({
            "facility_key": c["id"],
            "name": c.get("name"),
            "city": c.get("city"),
            "state": c.get("state"),
            "pin": c.get("pin"),
            "distance_km": dist,
            "flags": {k: c.get(k) for k in [
                "has_icu","has_nicu","has_dialysis","has_oncology","has_emergency_trauma",
                "has_surgery_ot","has_oxygen_supply","has_ventilator","open_24x7_claim","part_time_staff_claim"
            ]},
            "citations": citations
        })
    if any(r["distance_km"] is not None for r in results):
        results.sort(key=lambda x: (x["distance_km"] is None, x["distance_km"] if x["distance_km"] is not None else 1e9))
    state["results"] = results[:8]
    return state

def node_answer(state: AgentState) -> AgentState:
    context = []
    for r in state["results"]:
        context.append({
            "facility_key": r["facility_key"],
            "name": r["name"],
            "location": f"{r['city']}, {r['state']} {r['pin']}",
            "distance_km": r["distance_km"],
            "flags": r["flags"],
            "citations": r["citations"],
        })
    prompt = f"""
You are a healthcare facility reasoning agent.
Answer the user question ONLY using the provided candidate facilities and their citations.
Every factual claim about a facility must be backed by at least one citation snippet.
User question:
{state["question"]}
Candidate facilities (JSON):
{context}
Return STRICT JSON with:
- answer: string (brief)
- results: array of objects with {{facility_key, name, distance_km, why_matches: [..], citations:[..]}}
Do NOT invent facilities. Do NOT add capabilities without citations.
"""
    text = llm_answer(prompt)
    state["final"] = {"raw": text, "candidates_used": context}
    return state

g = StateGraph(AgentState)
g.add_node("retrieve", node_retrieve)
g.add_node("rank_and_cite", node_rank_and_cite)
g.add_node("answer", node_answer)
g.set_entry_point("retrieve")
g.add_edge("retrieve", "rank_and_cite")
g.add_edge("rank_and_cite", "answer")
g.add_edge("answer", END)
app = g.compile()

print("✓ Old single-agent system compiled (for backward compatibility)")
````
